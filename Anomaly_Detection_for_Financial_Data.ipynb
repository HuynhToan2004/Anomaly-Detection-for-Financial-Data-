```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = '/content/drive/MyDrive/Tiền xử lí và xây dựng bộ dữ liệu/BaiTap_28 4/train_data.xlsx'
test = '/content/drive/MyDrive/Tiền xử lí và xây dựng bộ dữ liệu/BaiTap_28 4/test_data.xlsx'
sample = '/content/drive/MyDrive/Tiền xử lí và xây dựng bộ dữ liệu/BaiTap_28 4/sample_solution.csv'

train_data = pd.read_excel(train)
test_data = pd.read_excel(test)
sample_data = pd.read_csv(sample)

train_data = train_data.drop(['id', 'id.1', 'W_Avg', 'Lead_Own_old'], axis=1)
test_data = test_data.drop(['id', 'Usage', 'W_Avg', 'Lead_Own_old'], axis=1)
```

```python
nan_indices_train = train_data['Lead_Own'][train_data['Lead_Own'].isna()].index
nan_indices_test = test_data['Lead_Own'][test_data['Lead_Own'].isna()].index
```

```python
train_data.loc[nan_indices_train, 'Lead_Own'] = train_data.loc[nan_indices_train, 'Vision']
test_data.loc[nan_indices_test, 'Lead_Own'] = test_data.loc[nan_indices_test, 'Vision']
```

```python
train_data['Lead_Own'] = train_data['Lead_Own'].astype('int64')
test_data['Lead_Own'] = test_data['Lead_Own'].astype('int64')
```

```python
from datetime import datetime

train_data['time'] = pd.to_datetime(train_data['time'], format = "%Y-%m-%d %H:%M:%S%z")
test_data['time'] = pd.to_datetime(test_data['time'], format = "%Y-%m-%d %H:%M:%S%z")
```

```python
train_data['time'] = train_data['time'].astype('int64')
test_data['time'] = test_data['time'].astype('int64')
```

```python
cat_cols = ['company', 'fund']
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
```

```python
cat_encoded_train = encoder.fit_transform(train_data[cat_cols])

cat_encoded_train_df = pd.DataFrame(cat_encoded_train, columns = encoder.get_feature_names_out(cat_cols))

# train_final = pd.concat([train_data.drop(cat_cols, axis=1), cat_encoded_train_df], axis=1)
```

```python
cat_encoded_test = encoder.fit_transform(test_data[cat_cols])

cat_encoded_test_df = pd.DataFrame(cat_encoded_test, columns = encoder.get_feature_names_out(cat_cols))

# test_final = pd.concat([test_data.drop(cat_cols, axis=1), cat_encoded_test_df], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scale_cols = ['time', 'IRR', 'Reduction']

scaler = StandardScaler()
```

```python
scaled_train = scaler.fit_transform(train_data[scale_cols])

scaled_train_df = pd.DataFrame(scaled_train, columns = scaler.get_feature_names_out(scale_cols))

train_final = pd.concat([cat_encoded_train_df, scaled_train_df, train_data.drop(scale_cols + cat_cols, axis=1)], axis=1)
```

```python
scaled_test = scaler.fit_transform(test_data[scale_cols])

scaled_test_df = pd.DataFrame(scaled_test, columns = scaler.get_feature_names_out(scale_cols))

test_final = pd.concat([cat_encoded_test_df, scaled_test_df, test_data.drop(scale_cols + cat_cols, axis=1)], axis=1)
```
